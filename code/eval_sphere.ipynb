```python
# Visualize single test run performance
import pandas as pd
import plotly.express as px

dim = 8

performance_df = pd.read_csv("../data/sphere_function_data/performance.csv")

performance_df = performance_df[performance_df.crossover == "UX"].drop(columns=["crossover"])
performance_df = performance_df[performance_df.seed_quality == "median"].drop(columns=["seed_quality"])

mean_performance_df = performance_df.groupby(["percentage_of_optimal_genes", "seed_fitness", "generation", "dim"], as_index=False).median()


fig = px.line(
    mean_performance_df[mean_performance_df.dim ==dim],
    y="min_f",
    x="generation",
    color="percentage_of_optimal_genes",
    labels={
        "value": "fitness",
        "variable": "generation",
        "percentage_of_optimal_genes": "% optimal genes",
    },
    )
fig.update_layout(yaxis_type="log")
fig.write_image("../figures/single_run_case_study_performance.pdf")
fig.show()
```

```python
#Single run case study Impact

import pandas as pd
import plotly.express as px

gen = 50
dim = 8

ci_df = pd.read_csv("../data/sphere_function_data/counting_impact.csv")
only_last_gen_ci = ci_df[ci_df.generation == gen]
only_last_gen_ci = only_last_gen_ci[only_last_gen_ci.dim == dim]
only_last_gen_ci = only_last_gen_ci[only_last_gen_ci.crossover == "UX"]
only_last_gen_ci = only_last_gen_ci.rename(columns={"traceID_1": "seed_ind"})

#CI of Fitness = 9
only_last_gen_ci = only_last_gen_ci[only_last_gen_ci.seed_quality == "median"]
fig = px.box(
    only_last_gen_ci.drop(columns=['generation', 'run_number', 'seed_quality', 'seed_fitness', 'crossover', 'dim']),
    color="percentage_of_optimal_genes",
    width=800, height=600,
    labels={
        "value": "impact",
        "variable": "",
        "percentage_of_optimal_genes": "% optimal genes",
    },
    )

fig.write_image("../figures/single_run_case_study_impact.pdf")
fig.show()
```

```python
#Fitness in the initial pop

import pandas as pd
import plotly.express as px

gen = 1
dim = 8

fitness_df = pd.read_csv("../data/sphere_function_data/fitness.csv")
fitness_df = fitness_df[fitness_df.generation == gen]
fitness_df = fitness_df[fitness_df.dim == dim]
fitness_df = fitness_df[fitness_df.crossover == "UX"]
fitness_df = fitness_df[fitness_df.seed_quality == "median"]
fitness_df = fitness_df[fitness_df.percentage_of_optimal_genes == 0]
fitness_df = fitness_df[fitness_df.run_number == 0].drop(columns = ["generation", "dim", "crossover", "seed_quality", "seed_fitness", "run_number", "percentage_of_optimal_genes", "individual"])
fitness_df = fitness_df.reset_index(drop=True)
fitness_df.index = fitness_df.index + 1


fig = px.bar(
    fitness_df,
    x=fitness_df.index[1:],
    y=fitness_df.f[1:],
    width=800, height=600,
    labels={
        "y": "fitness",
        "x": "individual",
    },
    )

fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 2,
        dtick = 1
    )
)
fig.add_hline(y=fitness_df.f[1])

fig.write_image("../figures/single_run_case_initial_fitness.pdf")
fig.show()
```
